```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Load the data
students_df = pd.read_csv('/content/drive/MyDrive/6070Student information.csv')
connections_df = pd.read_csv('/content/drive/MyDrive/Draft_6070relationship.csv')

# Step 2: Create a NetworkX graph
G = nx.Graph()

# Step 3: Add nodes with demographic attributes
for _, row in students_df.iterrows():
    G.add_node(row['Student_ID'], age=row['Age'], gender=row['Gender'], major=row['Major'])
```

```python

for _, row in connections_df.iterrows():
    G.add_edge(row['Student_ID_1'], row['Student_ID_2'], connection_type=row['Connection_Type'])

degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)

clustering_coefficient = nx.clustering(G)
print("Clustering Coefficient:", clustering_coefficient)

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=1, alpha=0.7, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")
plt.title("Student Social Network")
plt.show()
```
